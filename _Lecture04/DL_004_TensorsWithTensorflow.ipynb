## Calculate with Tensors

```python
a = tf.constant([[1, 2], [3, 4]])                 
b = tf.add(a, 1)

print(type(b))
b.numpy()   
```

```python
# now a float array
c = tf.constant([[1.0, 2], [3, 4]])
c.device
```

Summary: <br>
**Eager execution**: TensorFlow runs operations immediately without building a graph, which allows direct access to results via .numpy(). <br>
**Data type promotion**: When mixing types like integers and floats, TensorFlow automatically converts them to a common type (e.g., float) to keep computations consistent.<br>
**Device placement**: Tensors are stored and computed on a specific device (CPU or GPU), which you can inspect using the .device attribute.

## Building a computational graph 

```python
import tensorflow as tf

# check this out with and without @tf.function
@tf.function
def add(x):
    print("Python print in add")
    return x + 1

print("First call of add:")
print(add(tf.constant(3)))

print("\nSecond call of add:")
print(add(tf.constant(3)))
```

What is happening here: This example demonstrates the difference between normal Python execution and TensorFlow’s graph execution using @tf.function.

When the function add(x) is decorated with @tf.function, TensorFlow traces the function on the first call and builds a computational graph; during this tracing step, the Python statement print("Python print in add") is executed once. On subsequent calls, TensorFlow reuses the compiled graph instead of re-running the Python code, so the print statement does not appear again, even though the function still returns the correct numerical result. Without @tf.function, the function runs in normal eager mode, meaning the print statement would execute every time the function is called.

## Unvectorized addition & tensorflow variables

```python
c = tf.Variable(tf.zeros(n))
t = time.time()
for i in range(n):
    c[i].assign(a[i] + b[i])
f'{time.time() - t:.5f} sec'  # https://realpython.com/python-f-strings/#option-1-formatting 
```

Vectorized addition

```python
t = time.time()
d = a + b
f'{time.time() - t:.5f} sec'
```

Observation using a vectorized addition is much faster than using a python for loop.

# Comparison with numpy

```python
nn = 10000000
na = np.ones(nn)
nb = np.ones(nn)
t = time.time()
nd = na + nb
f'{time.time() - t:.5f} sec'
```
