## Creating and viewing a tensor

```python
import numpy as np
import torch
x = torch.tensor([1, 2, 3])
print('Type:', type(x))
print('Shape:', x.shape)
print(x)
```

```python
# convert to numpy 
print(x.numpy())
print(type(x.numpy()))
```

## Calculate with Tensors

```python
a = torch.tensor([[1, 2], [3, 4]])                 
b = torch.add(a, 1)

print(type(b))
b.numpy()   
```

```python
# now a float array
c = torch.tensor([[1.0, 2], [3, 4]])
c.device
```

# Vector math with tensors

```python
n = 10000
a = torch.ones(n)
b = torch.ones(n)
a
```

```python
# this will not work, since a tensor by default is not modifyable
a[2]=5.5
a
```

## Unvectorized addition & tensorflow variables

```python
c = torch.zeros(n)
t = time.time()
for i in range(n):
    c[i]=(a[i] + b[i])
f'{time.time() - t:.5f} sec'  # https://realpython.com/python-f-strings/#option-1-formatting 
```

Vectorized addition

```python
t = time.time()
d = a + b
f'{time.time() - t:.5f} sec'
```

# Comparison with numpy

```python
nn = 100000000
na = np.ones(nn)
nb = np.ones(nn)
t = time.time()
nd = na + nb
f'{time.time() - t:.5f} sec'
```

```python
nn = 100000000
na = torch.ones(nn)
nb = torch.ones(nn)
t = time.time()
nd = na + nb
f'{time.time() - t:.5f} sec'
```
