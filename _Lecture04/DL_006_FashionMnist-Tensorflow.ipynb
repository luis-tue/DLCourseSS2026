# First Tensorflow Example (Fashion MNIST)

This Notebook is about detecting T-Shirts in Fashion MNIST. We limit it to T-Shirts to start with binary classification instead of multinomial classification.

Get the necessary files for the next session from https://www.kaggle.com/datasets/zalando-research/fashionmnist


## Preparing the data

```python
import pandas as pd
import numpy as np
import tensorflow as tf 
print(tf.__version__)
print(tf.keras.__version__)
```

```python
#read data with pandas
dataTrain = pd.read_csv('data/fashion-mnist_train.csv')
dataTest = pd.read_csv('data/fashion-mnist_test.csv')
```

```python
# Look at the data, what can we see ? 
display(dataTrain)
#type(dataTrain)
```

```python
#interger location based indexing
X = dataTrain.iloc[:, 1:].to_numpy() # cut of the label column
y = dataTrain.iloc[:, 0].to_numpy() 

#Look at the data
print("X Shape", X.shape)
print("y Shape", y.shape)
```

```python
#convert to numpy
X_train, y_train = dataTrain.iloc[:, 1:].to_numpy(), dataTrain.iloc[:, 0].to_numpy()
X_test, y_test = dataTest.iloc[:, 1:].to_numpy(), dataTest.iloc[:, 0].to_numpy()

print("Training-data", X_train.shape)
print("Test-data", X_test.shape)
```

```python
#Define Classnames (taken from docu kaagle website)
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
class_names[3]
```

### Plot some examples with plt.imshow()

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([]) 
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_train[i].reshape(28, 28), cmap="binary") # 
    plt.xlabel(class_names[y_train[i]])
plt.show()
```

### Relabel images to binary decision about t-shirt 

```python
np.sum((y_train == 0)*1)
```

```python
# get from classes to binary classification for t-shirts 
y_train = (y_train == 0)
print("Nonzero", np.count_nonzero(y_train)) # 6000 t-Shirts 10%
y_test = y_test == 0
print("Nonzero", np.count_nonzero(y_test)) # 1000 t-Shirts 10%

```

```python
y_train
```

# Now classify all T-Shirts

```python
# Import Tensorflow Layer Types 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
print("Done")
```

```python
# Create a neuronal network with tensorflow
model = Sequential()
model.add(Input((784,)))
model.add(Dense(100, activation="sigmoid", name="inputlayer")) ##wieviel Gewichte ? 78400
model.add(Dense(1, activation="sigmoid")) # Gewichte 100*1
metrics=[
        'accuracy',
        #'precision',
        #tf.keras.metrics.Precision(),
        #tf.keras.metrics.Recall()
        ]
model.compile(optimizer="sgd", loss="binary_crossentropy", metrics=metrics)

print(model)
```

```python
print(model.summary())
```

```python
X_train.shape
```

```python
model.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=500)

#bei 60000 samples die 20 * gezeigt werden müssen insgesamt 1200000 beispiele gezeigt werden
# bei einer Batchsize von 1000 brauchen wir also 1200 Iterationen
```

```python
eval = model.evaluate(X_test, y_test)
```

```python
# try interpreting the return values:
eval
```

# Now also plot a graph for the development of the loss

```python
# retrain now with storage of the history
import time
history = model.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=1000)
```

```python
print(history.history.keys())
```

```python
history.history['loss']
```

```python
import matplotlib.pyplot as plt
print(history.history.keys())
# Create subplots: 1 row, 2 columns
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Plot accuracy
axs[0].plot(history.history['accuracy'])
axs[0].set_title('Model Accuracy')
axs[0].set_ylabel('Accuracy')
axs[0].set_xlabel('Epoch')
axs[0].legend(['Accuracy on training data'], loc='upper left')

# Plot loss
axs[1].plot(history.history['loss'])
axs[1].set_title('Model Loss')
axs[1].set_ylabel('Loss')
axs[1].set_xlabel('Epoch')
axs[1].legend(['Loss'], loc='upper left')

# Adjust layout and show the figure
plt.tight_layout()
plt.show()
```
